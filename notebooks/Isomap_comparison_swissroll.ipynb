```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import Isomap 
```

```python
#Load swiss roll
X = sklearn.datasets.make_swiss_roll(n_samples=1000)[0]
```

```python
print(X.shape)
```

```python
#Swiss roll visualization

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2])
```

```python
# Use Isomap
```

```python
embedding = Isomap(n_components=2)
x_iso = embedding.fit(X)
geodesics = x_iso.dist_matrix_
```

```python
print(geodesics.shape)
```

```python
print(geodesics[0,:])
```

```python
x0 = 20
x1 = 40
print(geodesics[x0,x1])
```
